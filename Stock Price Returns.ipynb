```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\hemaa\Downloads\stock_data.csv')
data
```

```python
data['5 Day Moving Average']  = data.AA.rolling(5).mean()
data['30 Day Moving Average']  = data.AA.rolling(30).mean()
```

```python
data.set_index('date', inplace=True)
```

```python
MA_data = data[['AA','5 Day Moving Average','30 Day Moving Average']]
data['Signal'] = data['30 Day Moving Average']
MA_data = data[['AA','5 Day Moving Average','30 Day Moving Average', 'Signal']]
MA_data
```

```python
MA_data['Buy/Sell'] = MA_data['Signal']
MA_data
```

```python
MA_data['Return'] = MA_data['Buy/Sell']
MA_data
```

```python
class trackPointChange():
    def __init__(self):
        self.previousPoint = 'None'
        self.yesterdayPrice = np.nan
        self.hold = False
    
    def trade_signal(self, row):
        row[3] = np.nan
        if self.previousPoint != '5DMA' and row[1] > row[2]:
            row[3] = row[1]
            self.previousPoint = '5DMA'
        elif self.previousPoint != '30DMA' and row[2] > row[1]:
            row[3] = row[2]
            self.previousPoint = '30DMA'
        
    def buy_signal(self, row):
        row[4] = np.nan
        if row[3] == row[2] and row[2] > 0:
            row[4] = 1
        elif row[3] == row[1] and row[1] > 0:
            row[4] = -1
                
    def log_return(self, row):
        row[5] = np.nan
        if row[4] == 1:
            self.hold = True
        elif row[4] == -1:
            self.hold = False
        if self.hold == True:
            row[5] = np.log(row[0]/self.yesterdayPrice)
        elif self.hold == False and row[4] == -1:
            row[5] = np.log(row[0]/self.yesterdayPrice)
        self.yesterdayPrice = row[0]
    
    def process_row(self, row):
        self.trade_signal(row)
        self.buy_signal(row)
        self.log_return(row)
        return row
        
determine_point = trackPointChange();
MA_data = MA_data.apply(determine_point.process_row, axis=1)
MA_data
```

```python
MA_returns = MA_data[['AA', '5 Day Moving Average', '30 Day Moving Average', 'Return']]
MA_returns
```
